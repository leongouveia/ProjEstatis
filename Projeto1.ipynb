```python
import pandas as pd
import numpy as np 
import zipfile                                                                                                   
```

Resposta 1

Na 1, importei zipfile, coloquei compression para descompactar e concatenei os dados dos 3 anos, fiz separado para usar posteriormente.  

```python
dados2020 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/jdDLrQIf33xXSCe/download',encoding='ISO-8859-9',compression='zip', sep=';')
dados2019 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/kRBUylqz6DyQznN/download',encoding='ISO-8859-9',compression='zip', sep=';')
dados2018 = pd.read_csv('https://arquivos.prf.gov.br/arquivos/index.php/s/MaC6cieXSFACNWT/download',encoding='ISO-8859-9',compression='zip', sep=';')
dadosCon = pd.concat([dados2020, dados2019, dados2018])
dadosCon
```

```python
#C
print('em 2020, estas foram as cidades que mais ocorreram acidentes:')
dds20 = dados2020.groupby(['municipio']).size().reset_index(name='count')
dds2= dds20.sort_values(by='count', ascending=False)
dds2.head()
```

```python
print('em 2019, estas foram as cidades que mais ocorreram acidentes:')
ds19= dados2019.groupby(['municipio']).size().reset_index(name='count')
dds3 = ds19.sort_values(by='count', ascending=False)
dds3.head()
```

```python
print('em 2018, estas foram as cidades que mais ocorreram acidentes:')
ds18 = dados2018.groupby(['municipio']).size().reset_index(name='count')
dds1= ds18.sort_values(by='count', ascending=False)
dds1.head()

```

```python
#d]D
pb19 = dados2019[(dados2019.uf == 'PB') & (dados2019.feridos_graves >= 1)]
print('Teve', len(pb19),'acidentes com feridos graves na Paraiba em 2019' )
pb19.head()
```

Resposta 3

Na A, ja tinha concatenado na 1, peguei, procurei por uf no groupby, usei o size para ver o tamanho e o reset index para imprimir a coluna contagem, ranquiei e salvei csv, apresentei todos.
Na B, fiz igual, so que para salvar em csv fiz uma copia.

```python
#A
print("Ranking da quantidade acidente por estado para todos os anos: ")
rankin = dadosCon.groupby(['uf']).size().reset_index(name='contagem')
rankinn= rankin.sort_values(by='contagem', ascending=False)
rankinn.to_csv("rankingporestado.csv")
rankinn
```

```python
#B
print('Quantidades de acidentes por dia da semana')
rankin2 = dadosCon.groupby(['dia_semana']).size().reset_index(name='contagem')
rankinn2= rankin2.sort_values(by='contagem', ascending=False)
print(rankinn2)
```

```python
rankingpordiadasemana = rankinn2.copy()
rankingpordiadasemana.to_csv("rankingpordiadasemana.csv")
```

Na B, peguei cada coluna pessoas, ilesos e feridos graves, fiz a soma, depois fiz a media para cada proporção, arrendondei para 2 digitos pos a virgula, então printei, fiz em todos os anos.

```python
#A
rankin = dadosCon.groupby(['causa_acidente']).size().reset_index(name='contagem')
rankinn= rankin.sort_values(by='contagem', ascending=False)
print('Causa mais frequente foi:')
rankinn.head(1)
```

```python
#B
somPes20 = dados2020['pessoas'].sum()
somIle20 = dados2020['ilesos'].sum()
somFeg20 = dados2020['feridos_graves'].sum()
mediPe20 = somIle20 / somPes20 
arrend20 = round(mediPe20,2)
mediPfg20 = round(somFeg20/somPes20,2)
print('Em 2020, aproximadamente:')
print('A proporção entre pessoas ilesas e envolvidas nos acidentes é:', arrend20, "e")
print('a proporção entre pessoas feridas gravemente e envolvidas nos acidentes é:', mediPfg20, ",ambos, a cada acidente.")
print()
print('Em 2019, aproximadamente:')
somPes19 = dados2019['pessoas'].sum()
somIle19 = dados2019['ilesos'].sum()
somFeg19 = dados2019['feridos_graves'].sum()
mediPe19 = round(somIle19 / somPes19,2)
mediPfg19 = round(somFeg19/somPes19,2)
print('A proporção entre pessoas ilesas e envolvidas nos acidentes é:', mediPe19, "e")
print('a proporção entre pessoas feridas gravemente e envolvidas nos acidentes é:', mediPfg19, ",ambos, a cada acidente.")
print()
print('Em 2018, aproximadamente:')
somPes18 = dados2018['pessoas'].sum()
somIle18 = dados2018['ilesos'].sum()
somFeg18 = dados2018['feridos_graves'].sum()
mediPe18 = round(somIle18 / somPes18,2)
mediPfg18 = round(somFeg18/somPes18,2)
print('A proporção entre pessoas ilesas e envolvidas nos acidentes é :', mediPe18, "e")
print('a proporção entre pessoas feridas gravemente e envolvidas nos acidentes é:', mediPfg18,",ambos, a cada acidente." )
```

```python
#C
dado20 = dados2020.copy()
dad20 = dado20.assign(Ano = 2020)
dado19 = dados2019.copy()
dad19 = dado19.assign(Ano = 2019)
dado18 = dados2018.copy()
dad18 = dado18.assign(Ano = 2018)
dadcon= pd.concat([dad20, dad19, dad18])
rankin = dadcon.groupby(['Ano']).size().reset_index(name='contagem')
rankinn= rankin.sort_values(by='contagem', ascending=False)
print("Aqui esta a contagem de acidentes por ano, provando que durante a pandemia a incidencia de acidente diminuiu.")
rankinn
```
